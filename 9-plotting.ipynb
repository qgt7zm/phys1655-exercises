# Plotting in Python

## Creating Simple Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# One dataset

x = [1, 2, 4, 7]
plt.plot(x)
plt.show()
```

```python
# Two datasets

x = [1,2,3,4,5]
y = [3,5,7,10,13]
plt.axis([0,6,2,14])
plt.plot(x,y,'ro')  # use red color and circle marker
plt.show()
```

## Customizing Plots

```python
def f(x):
    return np.exp(-x) * np.sin(2 * np.pi * x)

n = 100
x = np.linspace(0, 4, n+1)
y = f(x)

plt.plot(x, y, label='e^-x * sin(2πx)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axis([0, 4, -0.5, 0.8])
plt.title('My First Matplotlib Demo')

plt.savefig('fig.png')
```

## Multiple Graphs per Plot

```python
x = np.arange(1,19,.4)
y1 = np.log10(x)
y2 = 0.01 * x**2
y3 = 0.9 * np.sin(x)

# Plot datasets before showing
plt.plot(x,y1,'r-',label='y1')
plt.plot(x,y2,'b^',label='y2')
plt.plot(x,y3,'go',label='y3')
plt.plot(x,y1 + y2 + y3,'+', label='y1 + y2 + y3')

plt.legend(loc=2)
plt.show()
```

## Error Bars

```python
# Generate random data and errors
x = np.arange(10) + 2 * np.random.randn(10)
y = np.arange(10) + 2 * np.random.randn(10)

xerr = 2 * np.random.random(10)
yerr = 2 * np.random.random(10)

plt.errorbar(x,y,xerr=xerr,yerr=yerr,fmt='bo')
plt.show()
```

## Multiple Plots per Page

```python
def f(x):
    return np.sin(2*np.pi*x)

x1 = np.arange(0.0,4.0,0.1)
x2 = np.arange(0.0,4.0,0.01)
y1 = f(x1)
y2 = f(x2)

y1n = y1 + 0.1 * np.random.randn(len(x1))

plt.figure()  # Create new figure

plt.subplot(2, 1, 1)  # Rows, columns, index
plt.plot(x1,y1,'bo',x2,y2,'r:')

plt.subplot(2, 1, 2)
plt.plot(x1,y1n,'bo',x2,y2,'r:')

plt.show()
```

## Histograms

```python
# Generate data
x = np.random.randn(500)

# Generate bins
width = 0.5
histmin = np.floor(min(x))
histmax = np.ceil(max(x)) + width
bins = np.arange(histmin, histmax, width)

plt.hist(x,bins=bins)
plt.show()
```

## Piecewise Functions

```python
def H(x):
    return np.where(x < 0, 0.0, 1.0)  # vectorized boolean

n = 10
x = np.linspace(-5, 5, n+1)
y = H(x)
plt.plot(x, y)
```
